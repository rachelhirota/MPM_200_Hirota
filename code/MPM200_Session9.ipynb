```python
birds_meta.rename(columns = {"2024 IUCN Red List category":"Red List status"}, inplace = True)

mammals_meta["Scientific name"] = mammalia_mammals_meta["Genus"] + mammalia_mammals_meta["Species"]
mammals_meta.rename(columns = {"Common names (Eng)":"Common name"}, inplace = True)
```

```python
mammals_meta = mammals_meta[mammals_meta["Class"]=="MAMMALIA"]
df_mammals = mammals_meta[["Order", "Family", "Scientific name", "Common name", "Authority", "Red List status"]]

birds_meta = birds_meta[birds_meta["Subsp. Seq."]==0.0]
df_birds = birds_meta[["Order", "Family", "Scientific name", "Common name", "Authority", "Red List status"]]

metadata = pd.concat([df_mammals, df_birds])
```

```python
print(df_mammals.shape)
print(df_birds.shape)

metadata.shape
#16830x6
```

```python
df.rename(columns = {"Scientific Name":"Scientific name"}, inplace = True)

df_inner = pd.merge(df, metadata, on="Scientific name", how="inner")
df_outer = pd.merge(df, metadata, on="Scientific name", how="outer")
df_left = pd.merge(df, metadata, on="Scientific name", how="left")
df_right = pd.merge(df, metadata, on="Scientific name", how="right")
```
